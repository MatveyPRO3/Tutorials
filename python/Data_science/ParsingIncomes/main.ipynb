Processing data and creating dataframes

```python
import pandas as pd
import plotly.express as px
import numpy as np
from statistics import median
import csv

#  global vars

countries = ["Netherlands","Montenegro","United Kingdom","Portugal","Norway"]
render_data = {"country":[],"incomes":[],"year":[],"age":[]}


with open("DataSets/ilc_di03_1_Data.csv", "r") as file:
    raw_data = csv.reader(file)
    for row in raw_data:
        if row[0] == "TIME": #  Skipping first line
            continue
        if (row[5] == "Euro") and (row[1] in countries or row[1][:14] == "European Union"):
            render_data["country"].append(row[1][:14])
            if row[6] == ":":
                render_data["incomes"].append(None)
            else:
                render_data["incomes"].append(int(row[6].replace(",","")))
            render_data["year"].append(row[0])
            render_data["age"].append(row[2])
    

    df = pd.DataFrame.from_dict(render_data)
    df.to_csv("csv_tables/data_for_box_plots.csv")

    
    
file.close()
print("\033[36m Dataframes created successfully ✔")
```

```python
fig = px.box(df,x="country",y="incomes",points="all",notched=True,color="country",animation_frame="year")
fig.show(renderer='notebook_connected')
```
